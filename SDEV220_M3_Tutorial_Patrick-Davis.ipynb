Sort an array of 0s, 1s, and 2s.

```python
class Solution:
    # This is an algorithm called the Dutch national flag. It is commonly used
    # in problems like these.
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1
    
        while mid <= high:
            if arr[mid] == 0:
                # Swap arr[low] and arr[mid]
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:  # arr[mid] == 2
                # Swap arr[mid] and arr[high]
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

# Now we can use the function
array = [0,1,1,2,0,0,2,1,2]
solution = Solution() # Creates an instance of Solution class
solution.sort012(array) # Calls the method inside the class
print(array)
```

Binary search

```python
class Solution:
    def binarysearch(self, arr, k):
        count = -1
        for i in arr:
            if i == k:
                count += 1
                break
            elif i != k:
                count += 1
            else:
                count = -1
        return count

array = [1, 2, 3, 4, 5, 6]
k = 5
solution = Solution()
print(solution.binarysearch(array, k))
```
